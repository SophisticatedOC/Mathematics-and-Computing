```python
import numpy as np
import matplotlib.pyplot as plt
import math
import matplotlib as mpl


#sets font size
mpl.rcParams['font.size']=14

#generates x values from 0 to 10000000000 (10 zeros) and 1000 steps
x=np.linspace(0,1e10,10000)

#does the necessary calculations for the graph data
y_num=x+1
y_den=x+1e6
y=y_num/y_den

#creates the graphs plot lines
plt.plot(x,y_num,'b-')
plt.plot(x,y_den,'r--')
plt.plot(x,y,'g--')

#displays the legend, labels between the $ signs are in latex. the "r" prefix indicates python to expect latex language for the legend text
plt.legend([r'$x+1$',r'$x+10^6',r'$.frac{x+1}{x+10^6}$'])

#displays the axis labels
plt.xlabel('x')
plt.ylabel('x')



#task 1a


# Generate values for x, avoiding 0 to prevent division by zero
x = np.logspace(-10, 10, 500)  # Values between 1e-10 and 1e10

# Define numerator, denominator, and fraction
numerator = 1e-10
denominator = x**2
fraction = numerator / denominator

# Plot the three components
plt.figure(figsize=(12, 6))

# Plot numerator (constant line)
plt.plot(x, np.full_like(x, numerator), label='Numerator (1e-10)', linestyle='--', color='red')

# Plot denominator (x^2)
plt.plot(x, denominator, label='Denominator (x²)', linestyle='-', color='blue')

# Plot fraction (1e-10 / x²)
plt.plot(x, fraction, label='Fraction (1e-10 / x²)', linestyle='-', color='green')

# Log scale for better visualization
plt.xscale('log')
plt.yscale('log')

# Labels and title
plt.xlabel('x')
plt.ylabel('Value')
plt.title('Numerator, Denominator, and Fraction')
plt.legend()
plt.grid(True, which="both", ls="--")

# Displays the plot
plt.show()



#exponential


x=np.linspace(0,5,1000)
y=np.exp(x)
plt.plot(x,y,'b-')
plt.plot(x,x**2,'g--')
plt.plotx(x,x,'r--')
plt.legend([r'$y=e^x$',r'$y=x^2$',r'$y=x$'])
plt.xlabel('x');plt.ylabel('y')



#exponential quickly exceeds machine precision


x=np.linspace(0,100,1000)
y=np.exp(x)
plt.plot(x,y,'b-')
plt.plot(x,x**2,'g--')
plt.plot(x,x**10,'r--')
plt.legend([r'$y=e^x$',r'$y=x^2$',r'$y=x^[10]$'])
plt.xlabel('x');plt.ylabe('y')

```
